```python
import pandas as pd
from app import calibrateTargets
```

```python
# Cargar el dataset
path_data = '../data/pls_abstract_pairs_with_metrics.csv'
df = pd.read_csv(path_data)

# Separar train y test
train_df = df[df['split'] == 'train']
test_df = df[df['split'] == 'test']

# Preparar para fine-tuning (formato básico)
train_pairs = [
    {
        'original': row['source_text'],
        'humano': row['target_text']
    }
    for _, row in train_df.iterrows()
]

print("total de parejas de texto en entrenamiento: ", len(train_pairs))
```

```python
res_calibracion = calibrateTargets(
    path = path_data,
    save_to = "targets.json",
    subset = 0.1,          # % del dataset
    chunk_size = 16, 
)
```
