```python
import numpy as np
import matplotlib.pyplot as plt

def k_from_sigma_log(sigma, y_at_1sigma=0.95):
    return 1/sigma

def loss_centered(x, mu, sigma, y_at_1sigma=0.95):
    k = k_from_sigma_log(sigma, y_at_1sigma)
    s = 1/(1 + np.exp(-k*(x-mu)))
    return 2*np.abs(s - 0.5)

# Config
mus = [13.56, 14.96, 11.636]
sigmas = [2.08, 1.687, 0.79]
labels = ['FKGL','SMOG','Dalle Chall']

# Plot
plt.figure(figsize=(8,5))
for mu,sigma,lab in zip(mus,sigmas,labels):
    #for sigma in sigmas:
    x = np.linspace(mu - 5*sigma, mu + 5*sigma, 400)
    y = loss_centered(x, mu, sigma)
    plt.plot(x, y, label=lab)

plt.title('Prdida sigmoide centrada en $\\mu$ con $k=1/\\sigma$')
plt.xlabel('x')
plt.ylabel('loss(x)')
plt.ylim(-0.05, 1.05)
plt.grid(alpha=0.3)
plt.legend(loc='best')
plt.show()
```
